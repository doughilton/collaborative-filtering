```python
def find_missing_songs(user_playlist):
    # 1a
    spotify_usersongs = spotify_clean[spotify_clean['artist-song'].isin(user_playlist)]
    
    # 1b
    spotify_usersongs_playlistnames = spotify_clean[spotify_clean['artist-song'].isin(user_playlist)]['user-playlist'].unique()

    spotify_usersongs_full_playlists = spotify_clean[spotify_clean['user-playlist'].isin(spotify_usersongs_playlistnames)]
    
    playlist_scores = spotify_usersongs_full_playlists[spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('user-playlist')['artist-song'].count()
    playlist_scores = playlist_scores.rename('match_score')

    spotify_usersongs_full_playlists = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].merge(playlist_scores, left_on='user-playlist', right_index=True)

    song_scores = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('artist-song')['match_score'].sum()
    
    return song_scores.sort_values(ascending=False).head(10)
    
```

# Add penalization for playlist length

```python
spotify_usersongs = spotify_clean[spotify_clean['artist-song'].isin(user_playlist)]

spotify_usersongs_playlistnames = spotify_clean[spotify_clean['artist-song'].isin(user_playlist)]['user-playlist'].unique()

spotify_usersongs_full_playlists = spotify_clean[spotify_clean['user-playlist'].isin(spotify_usersongs_playlistnames)]

playlist_scores = spotify_usersongs_full_playlists[spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('user-playlist')['artist-song'].count()
playlist_scores = playlist_scores.rename('match_score')


## Penalization ##
playlist_penalization = (spotify_usersongs_full_playlists.groupby('user-playlist')['artist-song'].count()) ** 0.33
playlist_penalization = playlist_penalization.rename('match_score')

playlist_scores = playlist_scores / playlist_penalization
## Penalization ##


spotify_usersongs_full_playlists = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].merge(playlist_scores, left_on='user-playlist', right_index=True)

song_scores = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('artist-song')['match_score'].sum()

song_scores.sort_values(ascending=False).head(10)
```

# Penalization for song occurance across playlists

```python
spotify_usersongs = spotify_clean_cap[spotify_clean_cap['artist-song'].isin(user_playlist)]

spotify_usersongs_playlistnames = spotify_clean_cap[spotify_clean_cap['artist-song'].isin(user_playlist)]['user-playlist'].unique()

spotify_usersongs_full_playlists = spotify_clean_cap[spotify_clean_cap['user-playlist'].isin(spotify_usersongs_playlistnames)]

playlist_scores = spotify_usersongs_full_playlists[spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('user-playlist')['artist-song'].count()
playlist_scores = playlist_scores.rename('match_score')


## Playlist Penalization ##
playlist_penalization = (spotify_usersongs_full_playlists.groupby('user-playlist')['artist-song'].count()) ** 0.33
playlist_penalization = playlist_penalization.rename('match_score')

playlist_scores = playlist_scores / playlist_penalization
## Playlist Penalization ##

## Song Penalization ##
song_penalization = (spotify_usersongs_full_playlists.groupby('artist-song')['user-playlist'].count()) ** 0.33
song_penalization = song_penalization.rename('match_score')
## Song Penalization ##


spotify_usersongs_full_playlists = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].merge(playlist_scores, left_on='user-playlist', right_index=True)


# ## Sum of Playlist Score ##
# song_scores = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('artist-song')['match_score'].sum()
## Mean of Playlist Score ##
song_scores = spotify_usersongs_full_playlists[~spotify_usersongs_full_playlists['artist-song'].isin(user_playlist)].groupby('artist-song')['match_score'].sum() / len(playlist_scores)

song_scores = song_scores / song_penalization

song_scores.sort_values(ascending=False).head(10)
```
